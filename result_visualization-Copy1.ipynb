```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
dataset = "IHDP"
test_nb = 100

Nacq = 5

dir = "Sim/Single_Test" + str(test_nb)
Nsim = 200
acc = np.zeros((5, Nacq+1))
H = np.zeros((5, Nacq+1))


for filename in os.listdir(dir):
    if filename[0] == "a":
        acc_file = os.path.join(dir, filename)
        acc_tmp = np.load(acc_file)
        acc += acc_tmp
    elif filename[0] == "H":
        H_file = os.path.join(dir, filename)
        H_tmp = np.load(H_file)
        H += H_tmp
acc = acc/Nsim
H = H/Nsim
print(acc)
print(H)
```

```python
figure = plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(acc[0], color='blue', label='RS')
plt.plot(acc[1], color='orange', label='EIG/US')
plt.plot(acc[2], color='green', label='D-US')
plt.plot(acc[3], color='red', label='T-EIG')
plt.plot(acc[4], color='purple', label='D-EIG')
# plt.plot(acc[5], color='black', label='DM-Aware')
plt.legend()
plt.title("Accuracy")

plt.subplot(1,2,2)
plt.plot(H[0], color='blue', label='RS')
plt.plot(H[1], color='orange', label='EIG/US')
plt.plot(H[2], color='green', label='D-US')
plt.plot(H[3], color='red', label='T-EIG')
plt.plot(H[4], color='purple', label='D-EIG')
# plt.plot(H[5], color='black', label='DM-Aware')
plt.legend()
plt.title("Entropy")
plt.show()
```
